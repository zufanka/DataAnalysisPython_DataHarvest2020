```python
#open the zipfile containing the csvs and put it in the 'data' variable.
#(The ins and outs of this script are out of the scope of this session, sorry!)

zf = zipfile.ZipFile("DS-1180622.zip")
# zf.infolist()
data = pd.concat((pd.read_csv(zf.open(text_file.filename), encoding = "ISO-8859-1", na_values=":") for text_file in zf.infolist() if text_file.filename.endswith('Data.csv')))
len(data)
```

```python
# another errror: Unable to parse string "1,854" at position 289
# the values contain ',' and Pandas are therefore not recognizing them as numbers. Let's remove the commas and try again

pd.to_numeric(data["Value"].str.replace(",",""))
```

```python
# values at the beginning and the end of the data are missing, we can try and get a sample of 20 rows instead
pd.to_numeric(data["Value"].str.replace(",","")).sample(20)

# ok, it seems to be working.
```

```python
# let's re-assign the column values like this
data["Value"] = pd.to_numeric(data["Value"].str.replace(",",""))
```

```python
# Let's see how much masks was bought this year compared to previous year. The date information is in 'PERIOD'
data["PERIOD"]
# But this one too is an 'object'. We need it to be 'datetime'
```

```python
pd.to_datetime(data["PERIOD"])
```

```python
#  we can extract the year from a 'datetime' column like this
pd.to_datetime(data["PERIOD"]).dt.year
```

```python
# and assign it to a new column we will call "YEAR"
data["YEAR"] = pd.to_datetime(data["PERIOD"]).dt.year
```

```python
# let's check. Looks good!
data.head()
```

```python
# when we looked at the unique values per column above, we saw that we have multiple 'reporters' and 'partners' for the mask import.
# If we want a general indication of how much was spent on masks, we need to filter the data to prevent counting double
# So we will only keep the WORLD as from the PARTNER column. This is how we filter in pandas:
# give me only the rows from column PARTNER where the value is "WORLD", using '=='
data[data["PARTNER"] == "WORLD"]
```

```python
# all the countries are part of the EU27_2020 value in the "REPORTER" column. We want to exclude this one - because it does not contain any data anyway
# We can exclude parts of the dataset, using '!='

data[data["REPORTER"] != 'EU27_2020 (AT, BE, BG, CY, CZ, DE, DK, EE, ES, FI, FR, GR, HR,HU, IE, IT, LT, LU, LV, MT, NL, PL, PT, RO, SE, SI, SK)']
```

```python
# last, we only want the values, not quantity of the masks. We therefore filter
data[data["INDICATORS"] == "VALUE_IN_EUROS"]
```

```python
# This is how we merge all our filters.
# we will also assign them to a new variable called 'filtered_data'. This name is arbitrary, but chose variable names that you will understand

filtered_data = data[(data["INDICATORS"] == "VALUE_IN_EUROS") &
    (data["REPORTER"] != 'EU27_2020 (AT, BE, BG, CY, CZ, DE, DK, EE, ES, FI, FR, GR, HR,HU, IE, IT, LT, LU, LV, MT, NL, PL, PT, RO, SE, SI, SK)') &
    (data["PARTNER"] == "WORLD")]
```

```python
# to find out how much money was spent on imported masks, we will use .groupby() and sum up all the values. Only float and integer values will be summed up. This is how it works:
filtered_data.groupby("YEAR").sum()
```

```python
# this is the average price each country imported masks for
filtered_data.groupby("YEAR").mean()

# TIP: want to learn more about any method? Check the documentation: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html
```

```python
# after using .groupby(), your dataset will have the column(s) you group on set as index. This are the row names.
# To have the data accessible as a column again, use .reset_index()

filtered_data.groupby("YEAR").sum().reset_index()
```

```python
# let's visualize this using the altair library
import altair
```

```python
# here we need to use .reset_index() because altair can't access the index
per_year = filtered_data.groupby("YEAR").sum().reset_index()

# .Chart(data) : use this dataset
# .mark_bar() : make a bar chart
# .encode() : put details here
altair.Chart(per_year).mark_bar().encode(
    x = "Value",
    y = "YEAR:N" # 'N' tells altair to treat it as a string, not as a date. It's only so we have a prettier visualization!
)

# check out the altair example gallery for more code > https://altair-viz.github.io/gallery/index.html
```
