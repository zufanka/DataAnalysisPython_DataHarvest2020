```python
# when we looked at the unique values per column above, we saw that we have multiple 'reporters' and 'partners' for the mask import.
# If we want a general indication of how much was spent on masks, we need to filter the data to prevent counting double
# So we will only keep the WORLD as from the PARTNER column. This is how we filter in pandas:
# give me only the rows from column PARTNER where the value is "WORLD", using '=='
data[data["PARTNER"] == "WORLD"]
```

```python
# all the countries are part of the EU27_2020 value in the "REPORTER" column. We want to exclude this one - because it does not contain any data anyway
# We can exclude parts of the dataset, using '!='

data[data["REPORTER"] != 'EU27_2020 (AT, BE, BG, CY, CZ, DE, DK, EE, ES, FI, FR, GR, HR,HU, IE, IT, LT, LU, LV, MT, NL, PL, PT, RO, SE, SI, SK)']
```

```python
# last, we only want the values, not quantity of the masks. We therefore filter
data[data["INDICATORS"] == "VALUE_IN_EUROS"]
```

```python
# This is how we merge all our filters.
# we will also assign them to a new variable called 'filtered_data'. This name is arbitrary, but chose variable names that you will understand

filtered_data = data[(data["INDICATORS"] == "VALUE_IN_EUROS") &
    (data["REPORTER"] != 'EU27_2020 (AT, BE, BG, CY, CZ, DE, DK, EE, ES, FI, FR, GR, HR,HU, IE, IT, LT, LU, LV, MT, NL, PL, PT, RO, SE, SI, SK)') &
    (data["PARTNER"] == "WORLD")]
```

```python
filtered_data.head()
```

```python
# to find out how much money was spent on imported masks, we will use .groupby() and sum up all the values. Only float and integer values will be summed up. This is how it works:
filtered_data.groupby("YEAR").sum()
```

```python
# this is the average price each country imported masks for
filtered_data.groupby("YEAR").mean()

# TIP: want to learn more about any method? Check the documentation: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html
```

```python
# after using .groupby(), your dataset will have the column(s) you group on set as index. This are the row names.
# To have the data accessible as a column again, use .reset_index()

filtered_data.groupby("YEAR").sum().reset_index()
```

```python
# let's visualize this using the altair library
import altair
```

```python
# here we need to use .reset_index() because altair can't access the index
per_year = filtered_data.groupby("YEAR").sum().reset_index()

# .Chart(data) : use this dataset
# .mark_bar() : make a bar chart
# .encode() : put details here
altair.Chart(per_year).mark_bar().encode(
    x = "Value",
    y = "YEAR:N" # 'N' tells altair to treat it as a string, not as a date. It's only so we have a prettier visualization!
)

# check out the altair example gallery for more code > https://altair-viz.github.io/gallery/index.html
```

```python
# Want to go deeper?
# Group by multiple columns and explore!

slope = filtered_data.groupby(["REPORTER","YEAR"]).sum().reset_index()
slope.head()
```

```python
altair.Chart(slope).mark_line().encode(
    x = "YEAR",
    y = altair.Y("Value", scale=altair.Scale(type='log')),
    color = "REPORTER",
    tooltip = "REPORTER"
)
```

```python
# to calculate the difference between 2019 and 2020, we have to pivot the data this way
difference = pd.pivot_table(slope, index="REPORTER", columns="YEAR", values="Value").reset_index()
difference.head()
```

```python
difference.columns
```

```python
# difference_percentage is actually a misleading title here - the number is how many times as much the country paid for the masks in 2020 compared to 2019
difference["difference"] = difference[2020] / difference[2019]
difference.head()
```

```python
# if I only want to see particular columns, this is the syntax. Make sure you of keep the double [[]]
difference[["REPORTER", "difference"]]
```

```python
# let's compare the kilograms data to the value data. Did the countries that paid the most also get the most masks?
# I repeat all the steps I took for the VALUE_IN_EUROS
filtered_data_kg = data[(data["INDICATORS"] == "QUANTITY_IN_100KG") &
    (data["REPORTER"] != 'EU27_2020 (AT, BE, BG, CY, CZ, DE, DK, EE, ES, FI, FR, GR, HR,HU, IE, IT, LT, LU, LV, MT, NL, PL, PT, RO, SE, SI, SK)') &
    (data["PARTNER"] == "WORLD")]

difference_kg = pd.pivot_table(filtered_data_kg.groupby(["REPORTER","YEAR"]).sum().reset_index(), index="REPORTER", columns="YEAR", values="Value").reset_index()
difference_kg["difference"] = difference_kg[2020] / difference_kg[2019]
difference_kg.head()
```

```python
kg_chart = altair.Chart(difference_kg[["REPORTER", "difference"]]).mark_bar().encode(
    x = "difference",
    y = altair.Y("REPORTER", sort="-x")
)
```

```python
value_chart = altair.Chart(difference[["REPORTER", "difference"]]).mark_bar().encode(
    x = "difference",
    y = altair.Y("REPORTER", sort="-x")
)
```

```python
# we display these two next to one another
kg_chart | value_chart
```

```python
# let's see who got best and worst deals on masks and from whom

best_deal = data[(data["REPORTER"] != 'EU27_2020 (AT, BE, BG, CY, CZ, DE, DK, EE, ES, FI, FR, GR, HR,HU, IE, IT, LT, LU, LV, MT, NL, PL, PT, RO, SE, SI, SK)') &
    (data["PARTNER"] != "WORLD") &
    (data["YEAR"] == 2020) &
    ~(data["PARTNER"].str.contains("INTRA|EXTRA", na=False))] #using the ~ works similarly as != , it negates the statement
```

```python
best_deal.head()
```

```python
# we will group the data by reporter, partnera and indicators
best_deal_grouped = best_deal.groupby(["REPORTER","PARTNER","INDICATORS"]).sum().reset_index()
best_deal_grouped.head()
```

```python
# then we will pivot the table to calculate another column
best_deal_pivot = pd.pivot_table(best_deal_grouped,
                                 index=["REPORTER","PARTNER"],
                                 columns="INDICATORS",
                                 values="Value").reset_index()
best_deal_pivot.head()
```

```python
# I see that sometimes the QUANTITY_IN_100KG is 0 even if the VALUE_IN_EUROS is not. This is probably because it's not reported. For now, I will just remove the zero's
best_deal_pivot = best_deal_pivot[best_deal_pivot["QUANTITY_IN_100KG"] != 0]
best_deal_pivot.head()
```

```python
# and calculate the price per 100 kg
best_deal_pivot["price_p_100kg"] = best_deal_pivot["VALUE_IN_EUROS"] / best_deal_pivot["QUANTITY_IN_100KG"]
best_deal_pivot.head()
```

```python
# the PARTNER is a country code. I want the country name. I will therefore merge our data set with another one ...
countries = pd.read_csv("countries-iso2-iso3.csv")
countries.head()
```

```python
# ... using the pd.merge() function. In _on I specify what columns to match to one another
best_deal_merged = pd.merge(left=best_deal_pivot, right=countries, left_on="PARTNER", right_on="iso2")
best_deal_merged.head()
```

```python
# These are the 30 worst deals in the dataset. Greece got 700kg masks from Slovenia for 76,206 per 1000 kg
best_deal_merged[["REPORTER","QUANTITY_IN_100KG" ,"country", "price_p_100kg"]].nlargest(30, "price_p_100kg")
```

```python
# If we look at the statistics, this is a lot more than the 2,581.19 median or 4,116.12 mean
best_deal_merged.describe()
```

```python
# to save the table, I will export it as a csv
best_deal_merged[["REPORTER","QUANTITY_IN_100KG" ,"country", "price_p_100kg"]].nlargest(30, "price_p_100kg").to_csv("best_deals.csv", index=False)
```

```python
# you can combine the different methods used here in any way. 
# what is the average price per 100 kg of masks that came from Hong Kong?
# filter on Hong Kong and then .describe()
best_deal_merged[best_deal_merged["country"] == "Hong Kong"].describe()
```

```python
# what is the largest average price per 100 kg of masks?
# groupby by "country" (this is the country name of the PARTNER column), calculate the mean and filter out the largest values
best_deal_merged.groupby("country").mean().nlargest(10, "price_p_100kg")
```
